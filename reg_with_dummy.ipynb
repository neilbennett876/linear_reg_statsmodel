```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
sns.set()
```

```python
raw_data = pd.read_csv('1.03. Dummies.csv')
raw_data.describe()
```

```python
raw_data
```

```python
data = raw_data.copy()
```

```python
data['Attendance'] = data['Attendance'].map({'Yes':1,'No':0})
```

```python
data
```

```python
data.describe()
```

Regression

```python
y = data['GPA']
x1 = data[['SAT', 'Attendance']]
```

```python
x = sm.add_constant(x1) 
```

```python
results = sm.OLS(y,x).fit()
results.summary()
```

```python
plt.scatter(data['SAT'],y)
y_hat_no = .6439 + .0014*data['SAT']
y_hat_yes = .8665 + .0014*data['SAT']
fig = plt.plot(data['SAT'], y_hat_no, lw=4, c='red')
fig = plt.plot(data['SAT'], y_hat_yes, lw=4, c='blue')
plt.xlabel('SAT', fontsize=20)
plt.ylabel('GPA', fontsize=20)
plt.show()
```
